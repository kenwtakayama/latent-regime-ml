```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
mu = [
    [0, 0, 0, 0],
    [2, 2, -1, 1],
    [-2, 1, 2, -1]
]

Sigma = [
    [[1, 0.3, 0, 0],
     [0.3, 1, 0, 0],
     [0, 0, 1, 0.2],
     [0, 0, 0.2, 1]],

    [[1.5, -0.2, 0, 0],
     [-0.2, 1, 0.4, 0],
     [0, 0.4, 1, 0],
     [0, 0, 0, 1]],

    [[0.8, 0, 0.3, 0],
     [0, 1.2, 0, 0.4],
     [0.3, 0, 1, 0],
     [0, 0.4, 0, 1]]
]
```

```python
np.random.seed(42)
T_per_regime = 200
d = 4
X = []
z_true = []
```

```python
for k in range(3):
    samples = np.random.multivariate_normal(
        mean = mu[k],
        cov = Sigma[k],
        size = T_per_regime
        )
    X.append(samples)
    z_true.extend([k] * T_per_regime)

X = np.vstack(X)
z_true = np.array(z_true)
```

```python
fig, axes = plt.subplots(d, 1, figsize = (10, 8), sharex = True)

for i in range(d):
    axes[i].plot(X[:, i])
    axes[i].set_ylabel(f'x{i}')

plt.xlabel('Time')
plt.show()
```

```python
plt.figure(figsize=(10, 3))
plt.scatter(range(len(z_true)), z_true, s = 5)
plt.yticks([0, 1, 2])
plt.xlabel("Time")
plt.ylabel("True Regime")
plt.show()
```
