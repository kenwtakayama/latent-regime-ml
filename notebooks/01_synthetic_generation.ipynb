# Producing Synthetic Data

```python
import matplotlib.pyplot as plt
import sys, os
sys.path.append(os.path.abspath(".."))

from src.data_generation import generate_data, load_config
```

```python
# Load data from Configs

difficulty, config = load_config("medium")

X, z_true, d = generate_data(config)
```

```python
# Plot time-series features over time

fig, axes = plt.subplots(d, 1, figsize = (10, 10), sharex = True)

for i in range(d):
    axes[i].plot(X[:, i])
    axes[i].set_ylabel(f'x{i}')

fig.suptitle(f'Time-Series Features Over Time (Config: {difficulty.capitalize()})')
plt.xlabel('Time')
plt.savefig(f"../figures/01_Features_Over_Time_{difficulty.capitalize()}.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Plot true latent regimes over time

plt.figure(figsize=(10, 4))
plt.scatter(range(len(z_true)), z_true, s = 5)
plt.yticks([0, 1, 2])
plt.xlabel("Time")
plt.ylabel("True Regime")
plt.title(f"True Latent Regimes Over Time (Config: {difficulty.capitalize()})")
plt.savefig(f"../figures/01_True_Regimes_{difficulty.capitalize()}.png", dpi=300, bbox_inches='tight')
plt.show()
```
